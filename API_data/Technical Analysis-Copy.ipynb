```python
# Technical Analysis based Data

from pandas import read_csv
from pandas import datetime
import pandas as pd
```

```python
datafile =r'C:\Users\joseph\Documents\GitHub\crypto_trading\API_data\data\BTC_daily.csv'
```

```python
df = pd.read_csv(datafile,usecols=['timestamp','close'])
```

```python
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
df.index = df['timestamp']
```

```python
# Moving Averages

df['sma5'] = pd.Series(pd.rolling_mean(df['close'], 5))
df['sma7'] = pd.Series(pd.rolling_mean(df['close'], 7))
df['sma10'] = pd.Series(pd.rolling_mean(df['close'], 10))
df['sma30'] = pd.Series(pd.rolling_mean(df['close'], 30))
df['sma100'] = pd.Series(pd.rolling_mean(df['close'], 100))
df['sma200'] = pd.Series(pd.rolling_mean(df['close'], 200))
df['sma500'] = pd.Series(pd.rolling_mean(df['close'], 500))
df['sma1000'] = pd.Series(pd.rolling_mean(df['close'], 1000))
```

```python
import matplotlib.pyplot as plt
df1 = df[-100:]

plt.plot(df1['close'], 'g')
plt.plot(df1['sma5'], 'r')
plt.plot(df1['sma10'], 'b')
plt.show()
```
